```python
#brand map replacing the values in model

brandmap={
'华为':'Huawei',
'小米':'Xiaomi',
'三星':'Samsung',
'vivo':'vivo',
'OPPO':'OPPO',
'魅族':'Meizu',
'酷派':'Coolpad',
'乐视':'LeEco',
'联想':'Lenovo',
'HTC':'HTC'}

for key in brandmap:
    print(key + "is mapped to ", brandmap[key])
    model = model.replace([key], brandmap[key])
```

```python
#merging the datasets

temp_df = pd.merge(model, gender, on = 'device_id')
```
